# Galaxy Builder analysis code full walkthrough

In this notebook we demonstrate the basic usage of the `gzbuilder_analysis` package. While this code was originally designed to work with the main run of the [Galaxy Builder](https://www.zooniverse.org/projects/tingard/galaxy-builder/) citizen science project, much of the codebase is more general, including the rendering and fitting of oversampled, PSF-convolved Sersic profiles and clustering of poly-lines into logarithmic spirals.

In this notebook, we demonstrate
- Creating a synthetic galaxy image from a set of model parameters
- Parsing classifications of this synthetic image from the Zooniverse into a set of models
- Aggregating these models using unsupervised clustering and outlier detection
- Fitting this aggregate model using [JAX](jax.readthedocs.io), with a loss function that penalises deviation from the aggregate model

# Making a test-tube galaxy

The model we'll use contains a disk, bar and two spiral arms:

```python
with open('example_model.json') as f:
    model = pd.DataFrame(json.load(f)).unstack().dropna()
model
```

We also need to load some metadata, including a point-spread function (in the form of an array we convolve with the model), pixel mask, pixel uncertainties ("sigma image"), luminosity scaling multiplier and an image of the original galaxy which this synthetic one was based off of.

```python
with open('example_metadata.json') as f:
    metadata = pd.Series(json.load(f))
metadata['psf'] = np.array(metadata['psf'], dtype=float)
metadata[['sigma_image', 'galaxy_data']] = metadata[['sigma_image', 'galaxy_data']].apply(
    lambda a: np.ma.masked_array(a, ~np.array(metadata['pixel_mask'], dtype=bool), dtype=float)
)
```

Let's see what this model's isophotes look like:

```python
from descartes import PolygonPatch
from shapely.affinity import scale
from gzbuilder_analysis.aggregation import get_geoms
from gzbuilder_analysis.rendering import get_spirals

geoms = get_geoms(model)
f, ax = plt.subplots(figsize=(4, 4), dpi=100)
ax.add_patch(PolygonPatch(scale(geoms['disk'], 1, 1), fc='none', ec=f'C0'))
ax.add_patch(PolygonPatch(scale(geoms['bar'], 2, 2), fc='none', ec=f'C2'))
plt.xlim(30, 105)
plt.ylim(30, 105)
spirals = get_spirals(model.to_dict(), 2, model[('disk', 'roll')])
for s in spirals:
    plt.plot(*s.T, 'r')
plt.axis('off')
```

Let's render this model into an image! We can do this in a number of ways, either explicitly:

```python
def show_result(render):
    f, ax = plt.subplots(ncols=2, figsize=(8, 4))
    ax[0].imshow(stretch(render), cmap='gray', origin='lower')
    ax[1].imshow(stretch(metadata['galaxy_data']), cmap='gray', origin='lower')
    ax[0].set_title('Fake galaxy')
    ax[1].set_title('Real galaxy')
    [a.axis('off') for a in ax];
```

```python
from gzbuilder_analysis.rendering import oversampled_sersic_component, spiral_arm, psf_conv

disk = oversampled_sersic_component(model['disk'], metadata.galaxy_data.shape)
bar = oversampled_sersic_component(model['bar'], metadata.galaxy_data.shape)

spiral_points = get_spirals(model.to_dict(), 2, model[('disk', 'roll')])
spirals = []
for i in range(2):
    spirals.append(spiral_arm(
        arm_points=spiral_points[i],
        params={re.sub('\.[0-9]+', '', k): v for k, v in model['spiral'].items() if '.{}'.format(i) in k},
        disk=model['disk'],
        image_size=metadata.galaxy_data.shape,
    ))

noisy_render = np.array(psf_conv(disk + bar + sum(spirals), metadata['psf'])) \
    + np.random.randn(*metadata.galaxy_data.shape) * metadata.sigma_image

show_result(noisy_render)
```

Or using a helpful rendering function (note that this requires we define how many spiral arms are present):

```python
from gzbuilder_analysis.rendering import calculate_model

model[('spiral', 'n_arms')] = 2

rendered_model = calculate_model(model, image_size=metadata.galaxy_data.shape, psf=metadata['psf'])
noisy_render = rendered_model + np.random.randn(*metadata.galaxy_data.shape) * metadata.sigma_image

show_result(noisy_render)
```

Or, you can make use of the `Renderer` object (only for models where spirals are specified using log-spiral parameters, rather than as polylines). `Renderer` has the advantage of performing some check on the model, to ensure allowed parameters

```python
from gzbuilder_analysis.rendering.renderer import Renderer

r = Renderer(psf=metadata.psf, shape=metadata.galaxy_data.shape)
noisy_render = r(model) + np.random.randn(*metadata.galaxy_data.shape) * metadata.sigma_image
show_result(noisy_render)
```

It's also worth noting that none of these methods are as efficient as they could be. In the fitting section we'll use a faster method, which has been fully compiled using JAX (the above rely on JAX-compiled functions, but also use some Python logic to be more flexible).

## Bring in the Volunteers

Loading volunteer classifications from the Zooniverse is easy using pandas:

```python
classifications = pd.read_csv('example_classifications.csv', index_col=0)
classifications.head()
```

And we can quickly convert these into galaxy builder models:

```python
from gzbuilder_analysis.parsing import parse_classification

zoo_image_size = (512, 512)

zoo_models = classifications.apply(parse_classification, args=(zoo_image_size,), axis=1)
zoo_models.head().apply(lambda m: pd.DataFrame(m).unstack().dropna())
```

The Zooniverse images were scaled to 512x512 pixels, which we need to correct for:

```python
from gzbuilder_analysis.parsing import scale_model

scaled_models = zoo_models.apply(
    scale_model,
    args=(metadata.size_diff,),
)
```

We can also do some preliminary filtering, ensuring parameters are within allowed bounds

```python
from gzbuilder_analysis.parsing import sanitize_model

sanitized_models = scaled_models.apply(sanitize_model)
```

Let's look at the components provided by volunteers:

```python
from gzbuilder_analysis.aggregation.spirals import get_drawn_arms

f, ax = plt.subplots(nrows=2, ncols=2, figsize=(9, 8), dpi=70)
ax = ax.ravel()
[a.imshow(stretch(noisy_render), origin='lower', cmap='gray_r') for a in ax];
for geoms in sanitized_models.apply(get_geoms):
    if geoms.get('disk', None) is not None:
        ax[0].add_patch(PolygonPatch(scale(geoms['disk'], 1, 1), fc='none', ec='C0'))
    if geoms.get('bulge', None) is not None:
        ax[1].add_patch(PolygonPatch(scale(geoms['bulge'], 2, 2), fc='none', ec='C1'))
    if geoms.get('bar', None) is not None:
        ax[2].add_patch(PolygonPatch(scale(geoms['bar'], 2, 2), fc='none', ec='C2'))
        
for arm in get_drawn_arms(sanitized_models):
    ax[3].plot(*arm.T, 'r')
[a.axis('off') for a in ax]
ax[0].set_title('Drawn disks')
ax[1].set_title('Drawn bulges')
ax[2].set_title('Drawn bars')
ax[3].set_title('Drawn spiral arms')
plt.tight_layout();
```

```python
from gzbuilder_analysis.aggregation import AggregationResult

# provide the models and the target image (so that we can correctly scale clustering parameters)
agg_res = AggregationResult(sanitized_models, noisy_render)
```

```python
from gzbuilder_analysis.aggregation.spirals import get_drawn_arms

f, ax = plt.subplots(figsize=(5, 5))
plt.title('Aggregated Components')
ax.imshow(stretch(noisy_render), origin='lower', cmap='gray_r')
geoms = get_geoms(agg_res.model)
if geoms.get('disk', None) is not None:
    ax.add_patch(PolygonPatch(scale(geoms['disk'], 1, 1), fc='none', ec='C0'))
if geoms.get('bulge', None) is not None:
    ax.add_patch(PolygonPatch(scale(geoms['bulge'], 2, 2), fc='none', ec='C1'))
if geoms.get('bar', None) is not None:
    ax.add_patch(PolygonPatch(scale(geoms['bar'], 2, 2), fc='none', ec='C2'))    
for arm in get_spirals(agg_res.model.to_dict(), len(agg_res.spiral_arms), agg_res.model[('disk', 'roll')]):
    ax.plot(*arm.T, 'r')
ax.axis('off')
plt.tight_layout();
```

Volunteers have provided a bulge here, where one is not needed. This inclusion is a systematic problem across galaxy builder, but as you would expect most spiral galaxies to have some kind of bulge this is not a huge issue.

The next step is to fine-tune these parameters!

```python
from gzbuilder_analysis.fitting.optimizer import Optimizer
op = Optimizer(agg_res, *metadata[['psf', 'galaxy_data', 'sigma_image']])
```

This `Optimizer` object implements a well-optimized model rendering code (`gzbuilder_analysis.rendering.render_comps`) and loss calculation function, where the loss is a combination of the rendering residuals and the deviation from the initial parameters.

```python
%timeit op(op.model.values, op.model.index.values).block_until_ready()
```

Our initial guess is ignorant of the component brightnesses - we should fit it quickly

```python
from gzbuilder_analysis.fitting.misc import get_luminosity_keys

L_keys = get_luminosity_keys(op.model)
```

Define two handy functions to read results back from the GPU for scipy's LBFGS-b (and one to help with a progress bar)

```python
def __f(p, optimizer, keys):
    return np.array(optimizer(p, keys).block_until_ready(), dtype=np.float64)

def __j(p, optimizer, keys):
    return np.array(optimizer.jac(p, keys).block_until_ready(), dtype=np.float64)

def __bar_incrementer(bar):
    def f(*a, **k):
        bar.update(1)
    return f
```

And run the brightness fit:

```python
from tqdm import tqdm
from scipy.optimize import minimize

# perform the first fit
with tqdm(desc='Fitting brightness', leave=True) as bar:
    res = minimize(
        __f,
        np.array([op.model_[k] for k in L_keys]),
        jac=__j,
        args=(op, L_keys),
        callback=__bar_incrementer(bar),
        bounds=np.array([op.lims_[k] for k in L_keys]),
    )
```

Update the optimizer with the new parameters

```python
for k, v in zip(L_keys, res['x']):
    op[k] = v
```

```python
plt.imshow(stretch(psf_conv(sum(op.render_comps().values()), op.psf)), origin='lower', cmap='gray_r')
```

Perform the full fit, this will be much faster if you have a GPU and have installed JAX with CUDA!

```python
with tqdm(desc='Fitting everything', leave=True) as bar:
    res_full = minimize(
        __f,
        np.array([op.model_[k] for k in op.keys]),
        jac=__j,
        args=(op, op.keys),
        callback=__bar_incrementer(bar),
        bounds=np.array([op.lims_[k0][k1] for k0, k1 in op.keys]),
        options=dict(maxiter=500)
    )
```

```python
from copy import deepcopy
```

```python
final_model = pd.Series({
    **deepcopy(op.model_),
    **{k: v for k, v in zip(op.keys, res_full['x'])}
})
```

```python
from gzbuilder_analysis.fitting.misc import get_luminosity_keys, \
    remove_zero_brightness_components, lower_spiral_indices, correct_spirals, \
    correct_axratio
```

Correct the parameters of spirals in this model for the new disk, allowing rendering of the model without needing the rotation of the disk before fitting

```python
final_model = correct_spirals(final_model, op.base_roll)
```

Fix component axis ratios (if > 1, flip major and minor axis)

```python
final_model = correct_axratio(final_model)
```

Remove components with zero brightness

```python
final_model = remove_zero_brightness_components(final_model)
```

Lower the indices of spirals where possible

```python
final_model = lower_spiral_indices(final_model)
```

Let's render the fit galaxy using the `Optimizer` object and show the residuals:

```python
comps = op.render_comps(final_model.to_dict(), correct_spirals=False)
d = np.ma.masked_array(psf_conv(sum(comps.values()), op.psf) - op.target, op.mask)
lim = np.abs(d).max()
plt.imshow(d, vmin=-lim, vmax=lim, origin='lower', cmap='seismic')
plt.axis('off');
```

We can calculate the reduced chi-squared of the fit easily too

```python
from IPython.display import Latex
chisq = float(np.sum((d[~op.mask] / op.sigma[~op.mask])**2) / (~op.mask).sum())
Latex(r'$\chi_\nu^2 = {:.4f}$'.format(chisq))
```

```python
# fractions were originally parametrized vs the disk and spirals (bulge
# had no knowledge of bar and vice versa). We correct that here
disk_spiral_L = (
    final_model[('disk', 'L')]
    + (comps['spiral'].sum() if 'spiral' in comps else 0)
)

bulge_frac = final_model.get(('bulge', 'frac'), 0)
bar_frac = final_model.get(('bar', 'frac'), 0)

bulge_L = bulge_frac * disk_spiral_L / (1 - bulge_frac)
bar_L = bar_frac * disk_spiral_L / (1 - bar_frac)
gal_L = disk_spiral_L + bulge_L + bar_L

bulge_frac = bulge_L / (disk_spiral_L + bulge_L + bar_L)
bar_frac = bar_L / (disk_spiral_L + bulge_L + bar_L)

print(f'Bulge (bar) fraction: {bulge_frac:.4f} ({bar_frac:.4f})')
```

It is possible to estimate parameter errors using the local curvature of likelihood space, but we should be careful with this as these errors are likely to be severe underestimates

```python
ftol = 2.220446049250313e-09

hess_errs = np.sqrt(
    max(1, abs(res_full.fun))
    * ftol
    * np.diag(res_full.hess_inv.todense())
)
```
